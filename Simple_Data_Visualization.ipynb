```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Selecting the database from system
df = pd.read_csv(r'C:\Users\user\Desktop\smartphones.csv')
# Selecting relevant columns and dropping rows with missing values in RAM, Storage, and Final Price
smartphones_cleaned = df[['RAM', 'Storage', 'Final Price']].dropna()

# Scatter plot matrix to observe pairwise relationships
sns.pairplot(smartphones_cleaned)
plt.suptitle("Figure 1: Scatter Plot Matrix of RAM, Storage, and Final Price", y=1.02)
plt.show()

```

```python
top_brands = df['Brand'].value_counts().index[:5]  # Selecting top 5 brands for clearer visualization
smartphones_filtered = df[df['Brand'].isin(top_brands)].dropna(subset=['RAM', 'Storage', 'Final Price'])

# Trellis plot (FacetGrid) of RAM vs Final Price, colored by Storage
g = sns.FacetGrid(smartphones_filtered, col='Brand', col_wrap=3, height=4, sharex=False, sharey=False)
g.map_dataframe(sns.scatterplot, x='RAM', y='Final Price', hue='Storage', palette='viridis', edgecolor="w")
g.add_legend()
g.set_axis_labels("RAM (GB)", "Final Price (USD)")
g.fig.suptitle("Figure 2: Trellis Plot of RAM vs. Final Price by Brand", y=1.05)
g.fig.subplots_adjust(hspace=0.3)
for ax in g.axes.flat:
    ax.set_xlabel("RAM (GB)")


plt.show()
```

```python
# Calculate the correlation matrix for RAM, Storage, and Final Price
correlation_matrix = smartphones_cleaned.corr()

# Plot the correlation matrix as a heatmap
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Figure 3: Correlation Heatmap of RAM, Storage, and Final Price")
plt.show()

```
